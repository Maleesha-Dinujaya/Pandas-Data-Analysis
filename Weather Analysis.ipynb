```python
#Calculate and display the mean temperature for each state
mean_t = df.groupby ('State')['Temperature'].mean()
print(mean_t)
```

```python
#Find the maximum wind speed recorded and the corresponding date

max_wind_speed_index = df['Wind Speed'].idxmax()
max_wind_speed_date = df.loc[max_wind_speed_index, 'Date']
max_wind_speed_value = df.loc[max_wind_speed_index, 'Wind Speed']

print("Maximum Wind Speed Recorded:", max_wind_speed_value, "on", max_wind_speed_date)

```

```python
#Calculate the total precipitation for the entire dataset
total_precipitation = df['Data.Precipitation'].sum()
total_precipitation
```

```python
#Sort the DataFrame by temperature in ascending order
df_sort_by_temp = df.sort_values(by='Temperature')
df_sort_by_temp
```

```python
#Sort the DataFrame by precipitation in descending order
df_sort_by_prp = df.sort_values(by='Data.Precipitation',ascending  = False)
df_sort_by_prp
```

```python
#Create a line plot to visualize the temperature changes over time (using the 'Date' column)
import matplotlib.pyplot as plt
```

```python
df['Date'] = pd.to_datetime(df['Date'])

plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Temperature'], marker='o', linestyle='-', color='b')
plt.title('Temperature Changes Over Time')
plt.xlabel('Date')
plt.ylabel('Temperature')
plt.grid(True)
plt.show()
```

```python
average_temperature_by_state = df.groupby('State')['Temperature'].mean()

plt.figure(figsize=(10, 6))
average_temperature_by_state.plot(kind='bar', color='skyblue')
plt.title('Average Temperature by State')
plt.xlabel('State')
plt.ylabel('Average Temperature')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
#Check for missing values in the dataset

missing_values = df.isna().sum()
total_missing = missing_values.sum()
print("Missing values in each column:")
print(missing_values)
print("Total missing values in the DataFrame:", total_missing)
```
